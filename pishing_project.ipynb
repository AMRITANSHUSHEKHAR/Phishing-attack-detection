```python
# LOADING DATA
# importing packages to be used
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading data for dataset
data_url= pd.read_csv('urldata.csv')
data_url.head()
```

```python
# FAMILIARIZING WITH DATA
# checking the shape of dataset
data_url.shape
```

```python
# listing all the columns/features of the dataset
data_url.columns
```

```python
# getting the information of dataset
data_url.info()
```

```python
# VISUALIZING THE DATA
# plotting the data distribution

data_url.hist(bins=70, figsize= (15,15))
plt.show()
```

```python
# DATA PREPROCESSING AND EDA (Exploratory Data Analysis)
data_url.describe()
```

```python
# DROPPING THE DOMAIN COLUMN
new_data= data_url.drop(['Domain'], axis=1).copy()
```

```python
# checking the data for null or missing values
new_data.isnull().sum()
```

```python
# shuffling the rows in the dataset so that when splitting the train and test
new_data = new_data.sample(frac=1).reset_index(drop=True)
new_data.head()
```

```python
# SPLITTING THE DATA
# separating and assigning features and target columns to X and Y
Y = new_data['Label']
X = new_data.drop('Label', axis=1)      # drop label from the columns(1) 
X.shape, Y.shape
```

```python
# splitting the dataset into train and test set in 80:20 ratio
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.8, random_state = 12)
X_train.shape, X_test.shape
```

```python
# MACHINE LEARNING MODELS AND TRAINING
from sklearn.metrics import accuracy_score

#to store the model performance result
ML_model = []
train_accuracy = []
test_accuracy = []

# function to call for storing the results
def store_result(model,a,b):
    ML_model.append(model)
    train_accuracy.append(round(a,3))
    test_accuracy.append(round(b,3))
```

```python
# Decision Tree Model
from sklearn.tree import DecisionTreeClassifier

# instantiate the Model
decision_tree = DecisionTreeClassifier(max_depth = 5)

# fit the model
decision_tree.fit(X_train,Y_train)
```

```python
# predicting the value from the model for the samples
Y_test_d_tree = decision_tree.predict(X_test)
Y_train_d_tree = decision_tree.predict(X_train)
```

```python
# computing the accuracy of the model performance
acc_train_d_tree = accuracy_score(Y_train,Y_train_d_tree)
acc_test_d_tree = accuracy_score(Y_test,Y_test_d_tree)

print("Decission Tree: ")
print("Accuracy on training Data: {:.3f}".format(acc_train_d_tree))
print("Accuracy on test Data: {:.3f}".format(acc_test_d_tree))
```

```python
#checking the feature improtance in the model
plt.figure(figsize=(8,5))
n_features = X_train.shape[1]
plt.barh(range(n_features), decision_tree.feature_importances_, align='center')
plt.yticks(np.arange(n_features), X_train.columns)
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.show()
```

```python
#storing the results
store_result('Decision Tree', acc_train_d_tree, acc_test_d_tree)
```

```python
# Random Forest Model
from sklearn.ensemble import RandomForestClassifier

# instantiative the model
forest = RandomForestClassifier(max_depth=5)

# fit the model
forest.fit(X_train, Y_train)
```

```python
# predicting the target value from the model 
Y_test_forest = forest.predict(X_test)
Y_train_forest = forest.predict(X_train)
```

```python
# computing the accuracy of the model performance
acc_train_forest = accuracy_score(Y_train,Y_train_forest)
acc_test_forest = accuracy_score(Y_test,Y_test_forest)

print("Random forest: ")
print("Accuracy on training Data: {:.3f}".format(acc_train_forest))
print("Accuracy on test Data: {:.3f}".format(acc_test_forest))
     
```

```python
# checking the feature improtance in the model
plt.figure(figsize=(8,5))
n_features = X_train.shape[1]
plt.barh(range(n_features), forest.feature_importances_, align='center')
plt.yticks(np.arange(n_features), X_train.columns)
plt.xlabel("Feature importance")
plt.ylabel("Feature")
plt.show()
```

```python
# storing the results
store_result('Random Forest', acc_train_forest, acc_test_forest)
```

```python
# XGBoost Classification model
from xgboost import XGBClassifier

# instantiate the model
xgb = XGBClassifier(learning_rate=0.4, max_depth=5)

# fit the model
xgb.fit(X_train, Y_train)
```

```python
#predicting the target value from the model 
Y_test_xgb = xgb.predict(X_test)
Y_train_xgb = xgb.predict(X_train)
```

```python
# computing the accuracy of the model performance
acc_train_xgb = accuracy_score(Y_train,Y_train_xgb)
acc_test_xgb = accuracy_score(Y_test,Y_test_xgb)

print("XGBoost: ")
print("Accuracy on training Data: {:.3f}".format(acc_train_xgb))
print("Accuracy on test Data: {:.3f}".format(acc_test_xgb))
```

```python
# storing the results
store_result('XGBoost', acc_train_xgb, acc_test_xgb)
```

```python
#Support vector machine model
from sklearn.svm import SVC

# instantiate the model
svm = SVC(kernel='linear', C=1.0, random_state=12)
#fit the model
svm.fit(X_train, Y_train)
```

```python
# predicting the target value from the model
Y_test_svm = svm.predict(X_test)
Y_train_svm = svm.predict(X_train)
```

```python
# computing the accuracy of the model performance
acc_train_svm = accuracy_score(Y_train,Y_train_svm)
acc_test_svm = accuracy_score(Y_test,Y_test_svm)

print("SVM: ")
print("Accuracy on training Data: {:.3f}".format(acc_train_svm))
print("Accuracy on test Data: {:.3f}".format(acc_test_svm))
```

```python
# storing the results
store_result('SVM', acc_train_svm, acc_test_svm)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame to store the results
results = pd.DataFrame({
    "ML Model": ML_model,
    "Train Accuracy": train_accuracy,
    "Test Accuracy": test_accuracy
})

# Sort the DataFrame by test accuracy in descending order
results = results.sort_values(by='Test Accuracy', ascending=False)

# Plotting the bar graph
plt.figure(figsize=(7, 5))

plt.bar(results['ML Model'], results['Train Accuracy'], label='Train Accuracy', color='blue', alpha=0.7)
plt.bar(results['ML Model'], results['Test Accuracy'], label='Test Accuracy', color='red', alpha=0.7)

plt.xlabel('Machine Learning Model')
plt.ylabel('Accuracy')
plt.title('Comparison of Machine Learning Models')
plt.legend()
plt.ylim(0, 1)  # Set y-axis limit to ensure accuracy values are within the plot range
plt.show()
```

```python
# save XGBoost model to file
import pickle
pickle.dump(xgb, open("XGBClassifier.pickle.dat","wb"))
```

```python
# testing the saved model
# load model from file

xgb_model = pickle.load(open("XGBClassifier.pickle.dat","rb"))
predictions=xgb_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

accuracy = accuracy_score(Y_test, predictions)
report = classification_report(Y_test, predictions)
matrix = confusion_matrix(Y_test, predictions)

print("Accuracy: {:.3f} ".format(accuracy))
print("\nClassification Report:\n", report)
print("Confusion Matrix:\n", matrix)
```

```python
#Confusion Matrix Visualization

import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()


```
